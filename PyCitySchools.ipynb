```python
#school summary 
total_stu_sch = df2.groupby('school_name')['student_name'].count()
total_bud_sch = df1.groupby('school_name')['budget'].sum()

av_math_scr_sch = df2.groupby('school_name')['math_score'].sum() / total_stu_sch
av_rea_scr_sch = df2.groupby('school_name')['reading_score'].sum() / total_stu_sch

pass_math_sch = df2.loc[df2['math_score'] > 70, :].groupby('school_name')['math_score'].count()
pass_rea_sch =df2.loc[df2['reading_score'] > 70, :].groupby('school_name')['reading_score'].count()

schools_df_all = pd.concat([total_stu_sch, 
                                 total_bud_sch, 
                                 av_math_scr_sch, 
                                 av_rea_scr_sch, 
                                 pass_math_sch, 
                                 pass_rea_sch], 
                               axis = 1)

schools_df_all.reset_index(inplace=True)

schools_df_all = schools_df_all.rename(columns={'index' : 'student_name',  
                                                       'student_name' : 'Total Students',
                                                        'budget' : 'Total School Budget',
                                                        0 : 'Average Math Score',
                                                        1 : 'Average Reading Score',
                                                        'math_score' : 'passing math new',
                                                        'reading_score' : 'passing reading new'
                                                        })

schools_df_all = df1.merge(schools_df_all, on='school_name')

schools_df_all['Per Student Budget'] = schools_df_all['Total School Budget'] / schools_df_all['Total Students']
schools_df_all['% Passing Math'] = (schools_df_all['passing math new'] / schools_df_all['Total Students']) * 100
schools_df_all['% Passing Reading'] = (schools_df_all['passing reading new'] / schools_df_all['Total Students']) * 100
schools_df_all['% Overall Passing Rate'] = (schools_df_all['% Passing Math'] + schools_df_all['% Passing Reading']) / 2

schools_summary = schools_df_all[['school_name', 
                                       'type', 
                                       'Total Students', 
                                       'Total School Budget', 
                                       'Per Student Budget', 
                                       'Average Math Score', 
                                       'Average Reading Score', 
                                       '% Passing Math', 
                                       '% Passing Reading', 
                                       '% Overall Passing Rate']]
                                       
schools_summary = schools_summary.rename(columns={'type' : 'School Type'})
schools_summary.set_index('school_name', inplace=True)
schools_summary.index.name

schools_summary['Total School Budget'] = schools_summary['Total School Budget'].map('$ {:,.2f}'.format)
schools_summary['Per Student Budget'] = schools_summary['Per Student Budget'].map('$ {:,.2f}'.format)


schools_summary
```
